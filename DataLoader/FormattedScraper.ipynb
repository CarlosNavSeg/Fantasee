```python
import os
import pandas as pd
import json

# Definir la ruta base donde se encuentran los archivos JSON
base_path = '../marca-fantasy-api-scraper-updated-master/players'

# Lista para almacenar todos los DataFrames
dfs = []

# Recorrer cada carpeta dentro de la ruta base
for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)
    
    # Verificar que la ruta sea un directorio
    if os.path.isdir(folder_path):
        
        # Recorrer cada archivo JSON dentro de la carpeta
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.json'):
                file_path = os.path.join(folder_path, file_name)
                
                # Leer el archivo JSON y convertirlo a DataFrame
                with open(file_path, 'r') as f:
                    data = json.load(f)
                    df = pd.DataFrame(data['playerStats'])
                    df['playerTeam'] = file_path[file_path.rfind("/")+11:file_path.rfind("/")+15]
                    # Agregar una columna para identificar la fuente
                    df['source'] = file_path[file_path.rfind("_")+1:file_path.rfind(".")]
                    
                    # Agregar el DataFrame a la lista
                    dfs.append(df)

# Concatenar todos los DataFrames en uno solo
final_df = pd.concat(dfs, ignore_index=True)
# Mostrar el DataFrame final
final_df.head(5)
```

```python
def extraer_primer_elemento(lista):
    if isinstance(lista, list):  # Verificar si es una lista
        return lista[0]  # Devolver el primer elemento
    else:
        return 0  # Devolver 0 si no es una lista
```

```python
df = pd.DataFrame(final_df['stats'].tolist())
cleandf = df.iloc[:,:21]
for column in cleandf.columns:
    cleandf[column] = cleandf[column].apply(lambda x: extraer_primer_elemento(x))
```

```python
final_df = pd.concat([final_df, cleandf], axis = 1)
final_df.head()
```

```python
import re
def format_player_names(name):
    if(name.find('-') != -1 and bool(re.search(r'\d', name)) == False):
        playerName = name[name.find('-') + 1:].capitalize() +  ", " + (name[0:name.find('-')].capitalize())
        return playerName
    elif(bool(re.search(r'\d', name)) and name != 'de-frutos-1'):
        playerName = name[0:name.find('-')]
        return playerName.capitalize()
    elif(name == 'de-frutos-1'):
        return 'de Frutos, Jorge'
    else:
        return name.capitalize()
```

```python
final_df['playerName'] = final_df['playerName'].map(lambda x: format_player_names(x))
final_df['playerName'].unique()
```
