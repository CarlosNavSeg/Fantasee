```python
import pandas as pd
```

```python
df = pd.read_csv("playerStats.csv")
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
df['isInIdealFormation'] = df['isInIdealFormation'].map({False: 0, True: 1})
```

```python
df['isInIdealFormation']
```

```python
X = df.drop(['totalPoints', 'playerName', 'playerTeam', 'weekNumber'], axis = 1)
y = df['totalPoints']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
playerPredicted = df[(df['playerName'] == 'Gerard') & (df['weekNumber'] == 16)]
playerPredicted
```

```python
playerPredicted = playerPredicted.drop(['weekNumber', 'totalPoints', 'playerTeam', 'playerName'], axis = 1)
```

```python
gerardPrediction = svr.predict(playerPredicted)
gerardPrediction
```

```python
playerPredictionsBasedOnLastWeek = df[df["weekNumber"] == 24]
playerPredictionsBasedOnLastWeek.head(10)
```

```python
playerPredictions = playerPredictionsBasedOnLastWeek.drop(['weekNumber', 'totalPoints', 'playerTeam', 'playerName'], axis = 1)
```

```python
playerPredictionsScores = svr.predict(playerPredictions)
```

```python
import numpy as np
```

```python
playerPredictionsScores = np.around(playerPredictionsScores, decimals=0)
```

```python
df_predBasedOnLastWeek = pd.DataFrame({'playerName': playerPredictionsBasedOnLastWeek['playerName'],'playerTeam': playerPredictionsBasedOnLastWeek['playerTeam'], 'predicted_pointsBasedOnLastWeek': playerPredictionsScores})
df_predBasedOnLastWeek.head(10)
```

```python
promedio_por_jugador = df.iloc[:,4:26].groupby('playerName').agg('mean').reset_index()
promedio_por_jugador['isInIdealFormation'] = 0
ultima_columna = promedio_por_jugador.columns[-1]
nuevo_orden_columnas = [ultima_columna] + [col for col in promedio_por_jugador.columns if col != ultima_columna]
promedio_por_jugador = promedio_por_jugador.reindex(columns=nuevo_orden_columnas)
promedio_por_jugador
```
